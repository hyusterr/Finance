```python
from selenium import webdriver #注意：確認chromedriver.exe已經放到python環境的資料夾底下
from bs4 import BeautifulSoup as bs #沒有相應套件去terminal用 pip install 下載 
import pandas as pd
import time
```

```python
df = pd.DataFrame() #建立表格
browser = webdriver.Chrome() #開啟瀏覽器
browser.get('https://www.fsc.gov.tw/ch/home.jsp?id=131&parentpath=0,2&mcustomize=') #去網站

for i in range(1, 23):
    print('page' + str(i)) #看現在處理第幾頁

    if i <= 10:
        page = i 
        browser.find_element_by_xpath('//*[@id="messageform"]/div[5]/nav/ul/li['+ str(page) +']/a').click() #點按頁碼
        time.sleep(1)
        
        bigsoup = bs(browser.page_source, 'html.parser') #大碗湯
        souselist = bigsoup.findAll('div', {'class': 'souse1'}) #資料來源串列
        
        for event in range(6, 35, 2):
            row = pd.Series() #建立資料列
            print(int((event - 6) / 2) + 1)
            row['資料來源'] = souselist[int((event - 6) / 2)].text #填入資料來源
            
            browser.find_element_by_xpath('//*[@id="messageform"]/div[3]/div['+ str(event) +']/div[4]/a').click() #點進文章
            time.sleep(1)
            
            soup = bs(browser.page_source, 'html.parser') #小碗湯
            
            row['標題'] = (soup.findAll('h3')[-1].text) #填入標題、日期、內文
            row['發文日期'] = (soup.find('div', {'class': 'contentdate'}).text)
            row['內文'] = (soup.find('div', {'class':'page_content'}).text) #由於內文HTML格式不統一，只能連著日期一起抓
            
            df = df.append(row, ignore_index=True)
            
            browser.find_element_by_xpath('//*[@id="mform"]/div/a[6]/div').click()
            time.sleep(1)
    
        if i == 10: #換大頁
            browser.find_element_by_xpath('//*[@id="messageform"]/div[5]/nav/ul/li[11]/a').click()
     
    elif 11 <= i <= 20:
        page = i - 9
        browser.find_element_by_xpath('//*[@id="messageform"]/div[5]/nav/ul/li['+ str(page) +']/a').click() #點按頁碼
        
        bigsoup = bs(browser.page_source, 'html.parser')
        souselist = bigsoup.findAll('div', {'class': 'souse1'})
        for event in range(6, 35, 2):
            row = pd.Series()
            print(int((event - 6) / 2) + 1)
            row['資料來源'] = souselist[int((event - 6) / 2)].text
            
            browser.find_element_by_xpath('//*[@id="messageform"]/div[3]/div['+ str(event) +']/div[4]/a').click()
            time.sleep(1)
            
            soup = bs(browser.page_source, 'html.parser')
            row['標題'] = (soup.findAll('h3')[-1].text)
            row['發文日期'] = (soup.find('div', {'class': 'contentdate'}).text)
            row['內文'] = (soup.find('div', {'class':'page_content'}).text)
            
            df = df.append(row, ignore_index=True)
            
            browser.find_element_by_xpath('//*[@id="mform"]/div/a[6]/div').click()
            time.sleep(1)
            
        if i == 20:
            browser.find_element_by_xpath('//*[@id="messageform"]/div[5]/nav/ul/li[12]/a').click()
        
    elif 21 <= i <= 30:
    
        page = i - 19
        browser.find_element_by_xpath('//*[@id="messageform"]/div[5]/nav/ul/li['+ str(page) +']/a').click()#點按頁碼
        time.sleep(1)
        
        bigsoup = bs(browser.page_source, 'html.parser')
        souselist = bigsoup.findAll('div', {'class': 'souse1'})
        for event in range(6, 35, 2):
            row = pd.Series()
            print(int((event - 6) / 2) + 1)
            row['資料來源'] = souselist[int((event - 6) / 2)].text
            
            browser.find_element_by_xpath('//*[@id="messageform"]/div[3]/div['+ str(event) +']/div[4]/a').click()
            time.sleep(1)
            soup = bs(browser.page_source, 'html.parser')
            
            row['標題'] = (soup.findAll('h3')[-1].text)
            row['發文日期'] = (soup.find('div', {'class': 'contentdate'}).text)
            row['內文'] = (soup.find('div', {'class':'page_content'}).text)
            
            df = df.append(row, ignore_index=True)
            
            browser.find_element_by_xpath('//*[@id="mform"]/div/a[6]/div').click()
            time.sleep(1)

browser.close()
```

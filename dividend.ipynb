```python
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
from bs4 import BeautifulSoup
from selenium import webdriver
import random
import os
import shutil
import time
import pickle

pd.set_option('display.max_columns', None)
```

```python
# get all company id
with open('./uid.html', 'r') as f:
    html = f.read()
    f.close()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
uid_dict = {}
for span in soup.findAll('font'):
    if ' ' in span.text and span.text.split()[0].isdigit():
        uid_dict[ span.text.split()[0] ] = span.text.split()[1]
```

```python
done_list = [i.split('.')[0] for i in os.listdir('./div_files')]
final = []
```

```python
for com_code in done_list:
    
    print( 'dealing', com_code, '...' )
    
    tmp = pd.read_html( f'./div_files/{com_code}.html' )
    headers = tmp[0].iloc[-4:,:].values.tolist()
    tmp = tmp[0].iloc[:-4]
    tmp.columns = headers
    tmp['公司代號'] = pd.Series( [ com_code ] * len(tmp) )
    
    final.append( tmp )
    
    print( com_code, 'is done!')
```

```python
final_df = pd.concat( final, ignore_index=True)
```

```python
csv = pd.read_excel( 'dividend-policy.xlsx', header=[0, 1, 2, 3], index_col=0 )
csv = csv.rename(columns=lambda x: x if not 'Unnamed' in str(x) else '')
```

```python
fin = []

def str_to_float( string ):
    if string == '-':
        return None
    else:
        return float(string)


for company, data in csv.groupby(['公司']):
    print(company)
    # dividend
    tmp    = data[ data[('股 利 政 策', '股利發放年度', '股利發放年度', '股利發放年度')] != '∟' ]
    cumsum = tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '股利合計')][:-1][::-1].map(str_to_float).cumsum()[::-1]
    tail   = tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '股利合計')][-1:].map(str_to_float)
    tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '累計股利合計')] = pd.concat( [cumsum, tail] )
    
    try:
        tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '股利合計/累計股利合計')] = tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '股利合計')].map(str_to_float) / tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '累計股利合計')] 
    except ZeroDivisionError:
        tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '股利合計/累計股利合計')] = None
    
    # EPS
    cumsum_eps = tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', 'EPS(元)')][:-1][::-1].map(str_to_float).cumsum()[::-1]
    tail_eps   = tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', 'EPS(元)')][-1:].map(str_to_float)
    tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', '累計EPS(元)')] = pd.concat( [cumsum_eps, tail_eps] )
    tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', 'EPS/累計EPS(元)')] = tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', 'EPS(元)')].map(str_to_float) / tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', '累計EPS(元)')]
        
    tmp['累計股利/累計EPS'] = tmp[('股 利 政 策', '股東股利\xa0(元/股)', '股利合計', '累計股利合計')] / tmp[('盈餘分配率統計', '盈餘分配率統計', 'EPS(元)', '累計EPS(元)')]
    
    fin.append(tmp)
```

```python
pd.concat(fin).to_excel('dividend-policy-fin.xlsx')
```
